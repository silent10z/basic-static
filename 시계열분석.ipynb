시계열 분석(time series analysis)이란,
독립변수 를 이용하여 종속변수를 예측하는 일반적인 기계학습 방법론에 대하여 시간을 독립변수로 사용한다는 특징이 있다. 독립변수로 시간을 사용하는 특성때문에 분석에 있어서 일반적인 방법론들과 다른 몇가지 고려가 필요하다

ARIMA(Autoregressive Intefreated Moving Average)
자기회기모형, 이동평균모형
ARIMA는 시계열의 비정상성을 설명하기 위해 관측지간의 차분을 사용한다.

- AR:자기회기: 이전 관측값의 오차항 이후 관측값에 영향을 주는 모형, theta는 자기상관계수, epsilond은 white noise 이다. Time lag 1이상이 될수 있다.
- I: Intgrated. 누적을 의미하는 것으로 차분을 이용하는 시계열모형들에 붙이는 표현
- MA : 이동평균. 관측값이 이전 연속전인 오차항의 영향을 받는다는 모형이다
